```python
new_data = list()

for row in clean.to_dict(orient="records"):
    if " | " not in row["answers"] and "," in row["answers"] and "?," in row["questions"]:
        ans = row["answers"].split(",")
        qus = row["questions"].split("?,")
        
        for i in range(len(ans)):
            if qus[i][:-1] != "?":
                qus[i] = qus[i] + "?"
                
            new_item = row.copy()
            new_item["answers"] = ans[i]
            new_item["questions"] = qus[i]
            new_data.append(new_item)
        
    else:
        new_item = row.copy()
        new_data.append(new_item)
        
novel = pd.DataFrame.from_dict(new_data)
novel
```
